# Kaggle spaceship Titanic 
- goal : predict which passenger got transported to alternate dimension

# Data description
- train.csv - Personal records for about two-thirds (~8700) of the passengers, to be used as training data.
- PassengerId - A unique Id for each passenger. Each Id takes the form gggg_pp where gggg indicates a group the passenger is travelling with and pp is their number within the group. People in a group are often family members, but not always.
- HomePlanet - The planet the passenger departed from, typically their planet of permanent residence.
- CryoSleep - Indicates whether the passenger elected to be put into suspended animation for the duration of the voyage. Passengers in cryosleep are confined to their cabins.
- Cabin - The cabin number where the passenger is staying. Takes the form deck/num/side, where side can be either P for Port or S for Starboard.
- Destination - The planet the passenger will be debarking to.
- Age - The age of the passenger.
- VIP - Whether the passenger has paid for special VIP service during the voyage.
- RoomService, FoodCourt, ShoppingMall, Spa, VRDeck - Amount the passenger has billed at each of the Spaceship Titanic's many luxury amenities.
- Name - The first and last names of the passenger.
- Transported - Whether the passenger was transported to another dimension. This is the target, the column you are trying to predict.

Prior to exploring the content of categorical variables and their relationship, we create the core features:
- PassengerId: There is no overlapping bewteen groups. NO
- Cabin: deck/num/side
- Name: first name / last name

PassengerID
Spliting groups and id ?

```python
# no overlapping between groups of passenger -> no use for now
def group_id(df):
    groups  = df['PassengerId'].apply(lambda x: np.NaN if pd.isna(x) else x.split('_')[0])
    
    df['GroupId'] = groups
    return df

train  = group_id(train)
test   = group_id(test)
```

Cabin: splitting deck/num/side

```python
# splitting deck/num/side to separate features
def deck_num_side(df):
    try:
        deck = df['Cabin'].apply(lambda x: np.NaN if pd.isna(x) else x.split('/')[0])
        num = df['Cabin'].apply(lambda x: np.NaN if pd.isna(x) else x.split('/')[1])
        side = df['Cabin'].apply(lambda x: np.NaN if pd.isna(x) else x.split('/')[2])

        df['deck'] = deck
        df['num']  = num
        df['side'] = side

        return df.drop(columns = ['Cabin'])  
    except:
        print("Already processed")
        return df

train = deck_num_side(train)
test = deck_num_side(test)
```

```python
train.head()
```

Name: splitting surname/lastname

```python
# splitting surname and last name
def surname_lastname(df):
    try: 
        surname = df['Name'].apply(lambda x: np.NaN if pd.isna(x) else x.split(' ')[0])
        lastname = df['Name'].apply(lambda x: np.NaN if pd.isna(x) else x.split(' ')[1])

        df['Surname'] = surname
        df['lastname'] = lastname
        return df.drop(columns = ['Name'])
    except:
        print("Already processed")
        return df
    
train  = surname_lastname(train)
test   = surname_lastname(test) 
train.head()  
```

# III. Feature engineering
ideas:
- group_size
- family_size



```python
train.head()
```

```python
# create group_size feature

def make_group_size(df):

    # generate group:occurence map
    groups = df['GroupId'].value_counts().index
    occurence = df['GroupId'].value_counts().values
    group_map = {groups[ii] : occurence[ii] for ii in range(len(occurence))}
    # apply map to new columns
    df['GroupSize'] = np.NaN
    df['GroupSize'] = df['GroupId'].apply(lambda x: group_map.get(x) if x in group_map else np.NaN)
    
    return df

train = make_group_size(train)
test = make_group_size(test)
```

```python
# create family_size feature

def make_family_size(df):

    # generate group:occurence map
    groups = df['lastname'].value_counts().index
    occurence = df['lastname'].value_counts().values
    group_map = {groups[ii] : occurence[ii] for ii in range(len(occurence))}
    # apply map to new columns
    df['FamilySize'] = np.NaN
    df['FamilySize'] = df['lastname'].apply(lambda x: group_map.get(x) if x in group_map else np.NaN)
    
    return df

train = make_family_size(train)
test = make_family_size(test)
```

2. fill NaNs categories
Imputing missing values with random.choice from the actual set, keeping the distribution as is

```python
col_cat = ['HomePlanet', 'CryoSleep', 'Destination', 'VIP', 'deck', 'side', 'GroupSize']

def assign_random_values(df, cols):
    # this allows to preserve statistical distribution while imputing values:)
    df_out = df.copy()
    for col in cols:
        choice = df_out[col].dropna().values
        df_out[col] = df_out[col].apply(lambda x: x if not pd.isnull(x) else np.random.choice(choice))
    
    return df_out


train_3 = assign_random_values(train_2, col_cat)
test_3 =  assign_random_values(test_2, col_cat)
    
# display missing data
missingno.matrix(train_3, figsize = (30,5))
missingno.matrix(test_3, figsize = (30,5))
```

2. fill numericals
Imputing missing values with median

In the future we could fill values by fitting distribution and drawing from it :)

```python
col_num = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

def assign_random_values(df, cols):
    df_out = df.copy()
    for col in cols:
        # this allows to preserve statistical distribution while imputing values:)
        med = df_out[col].dropna().median()
        df_out[col] = df_out[col].apply(lambda x: x if not np.isnan(x) else med)
    
    return df_out

train_4 = assign_random_values(train_3, col_num)
test_4 =  assign_random_values(test_3, col_num)
    
    
# display missing data
missingno.matrix(train_4, figsize = (30,5))
missingno.matrix(test_4, figsize = (30,5))
```

# Model tuning

```python
# Import tuning model extensions
from sklearn.model_selection import GridSearchCV 
from sklearn.model_selection import RandomizedSearchCV 
```

```python
#simple performance reporting function
def clf_performance(classifier, model_name):
    print(model_name)
    print('Best Score: ' + str(classifier.best_score_))
    print('Best Parameters: ' + str(classifier.best_params_))
```

```python
%%time
# logistic regression
lr = LogisticRegression()
param_grid = {'max_iter' : [2000],
              'penalty' : ['l1', 'l2'],
              'C' : np.logspace(-4, 4, 20),
              'solver' : ['liblinear']}

clf_lr = GridSearchCV(lr, param_grid = param_grid, cv = 5, verbose = 1, n_jobs = -1)
best_clf_lr = clf_lr.fit(X_train,y_train)
clf_performance(best_clf_lr,'Logistic Regression')
# slight imporvement (1%)
```

```python
%%time
knn = KNeighborsClassifier()
param_grid = {'n_neighbors' : [3,5,7,9,12,14],
              'weights' : ['uniform', 'distance'],
              'algorithm' : ['auto', 'ball_tree','kd_tree'],
              'p' : [1,2]}
clf_knn = GridSearchCV(knn, param_grid = param_grid, cv = 5, verbose = True, n_jobs = -1)
best_clf_knn = clf_knn.fit(X_train,y_train)
clf_performance(best_clf_knn,'KNN')
# slight improvement (1%)
```

```python
%%time
svc = SVC(probability = True)
param_grid = tuned_parameters = [{'kernel': ['rbf'], 'gamma': [.1,0.2],
                                  'C': [.1, 10]},
                                 {'kernel': ['linear'], 'C': [.1, 0.2]}]
clf_svc = GridSearchCV(svc, param_grid = param_grid, cv = 5, verbose = True, n_jobs = -1)
best_clf_svc = clf_svc.fit(X_train,y_train)
clf_performance(best_clf_svc,'SVC')
# no improvement
```

```python
%%time
#Because the total feature space is so large, I used a randomized search to narrow down the paramters for the model. I took the best model from this and did a more granular search 
rf = RandomForestClassifier(random_state = 1)
param_grid =  {'n_estimators': [100,500,1000], 
                                  'bootstrap': [True,False],
                                  'max_depth': [3,5,10,20,75,100,None],
                                  'min_samples_leaf': [1,2,10],
                                  'min_samples_split': [2,5,10]}
                                  
clf_rf_rnd = RandomizedSearchCV(rf, param_distributions = param_grid, n_iter = 100, cv = 5, verbose = 3, n_jobs = -1)
best_clf_rf_rnd = clf_rf_rnd.fit(X_train,y_train)
clf_performance(best_clf_rf_rnd,'Random Forest')
```

```python
#Make Predictions 
voting_clf_hard.fit(X_train, y_train)
voting_clf_soft.fit(X_train, y_train)
voting_clf_all.fit(X_train, y_train)
voting_clf_xgb.fit(X_train, y_train)
best_rf.fit(X_train, y_train)

y_hat_vc_hard = voting_clf_hard.predict(X_test).astype(int)
y_hat_vc_soft =  voting_clf_soft.predict(X_test).astype(int)
y_hat_vc_all = voting_clf_all.predict(X_test).astype(int)
y_hat_vc_xgb = voting_clf_xgb.predict(X_test).astype(int)
y_hat_rf = best_rf.predict(X_test).astype(int)
```
